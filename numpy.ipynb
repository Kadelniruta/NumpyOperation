Statistical Analysis  
•  You are given a NumPy array of exam scores for 1000 students. Perform the following:  
o Calculate the mean, median, variance, and standard deviation of the scores.  
o Find the 25th, 50th, and 75th percentiles. 
o  Normalize the scores using min-max scaling and z-score normalization.  

```python
np.random.seed(0)
```

```python
scores = np.random.randint(0,101,size=1000)
print(scores)
```

```python
mean_score = np.mean(scores)
median_score = np.median(scores)
variance_score = np.var(scores)
std_deviation_score = np.std(scores)

print(f"Mean: {mean_score}")
print(f"Median: {median_score}")
print(f"Variance: {variance_score}")
print(f"Standard Deviation: {std_deviation_score}")

```

```python
percentile_25 = np.percentile(scores,25)
percentile_50 = np.percentile(scores,50)
percentile_75 = np.percentile(scores,75)
print(percentile_25)
print(percentile_50)
print(percentile_75)
```

```python
# Min-max normalization
min_scores = np.min(scores)
max_scores = np.max(scores)
normalized_scores = (scores - min_scores) / (max_scores - min_scores)
print(normalized_scores)
```

```python
# Z-score normalization
z_score_normalized_scores = (scores-mean_score)/std_deviation_score
print("Z-score normalized scores: ", z_score_normalized_scores)
```

7. Filtering Data  
•  Given a NumPy array representing a 2D dataset of height and weight of 200 individuals:  
o  Find all individuals with height > 180 cm and weight < 75 kg. o  
Calculate the 
mean and standard deviation of heights and weights for this subset. o  Create a 
mask to replace all weights greater than 100 kg with 100. 

```python

# Simulate a dataset with height (in cm) and weight (in kg) for 200 individuals
np.random.seed(42)  # Optional: for reproducibility
heights = np.random.randint(150, 201, 200)  # Heights between 150 and 200 cm
weights = np.random.randint(50, 151, 200)   # Weights between 50 and 150 kg
```

```python
# Combine heights and weights into a 2D array (each row is [height, weight])
dataset = np.column_stack((heights, weights))
print(dataset[:10])
```

```python

# 1. Find all individuals with height > 180 cm and weight < 75 kg

filtered_individuals = dataset[(dataset[:, 0] > 180) & (dataset[:, 1] < 75)]
print("\nIndividuals with height > 180 cm and weight < 75 kg:\n", filtered_individuals)


```

```python
# 2. Calculate the mean and standard deviation of heights and weights for this subset
mean_height = np.mean(filtered_individuals[:, 0])
mean_weight = np.mean(filtered_individuals[:, 1])
std_height = np.std(filtered_individuals[:, 0])
std_weight = np.std(filtered_individuals[:, 1])
print(f"Mean height: {mean_height:.2f} cm")
print(f"Mean weight: {mean_weight:.2f} kg")
print(f"Standard deviation of height: {std_height:.2f} cm")
print(f"Standard deviation of weight: {std_weight:.2f} kg")

```

```python

# 3. Create a mask to replace all weights greater than 100 kg with 100
weights_mask = np.where(dataset[:, 1] > 100, 100, dataset[:, 1])

# Update the original dataset's weight column with the modified weights
dataset[:, 1] = weights_mask

print("\nDataset after replacing weights > 100 kg with 100 (first 10 rows):\n", dataset[:10])
```

8. Time Series Simulation  
•  Simulate a random walk (cumulative sum of random steps) for 1000 steps, starting at 0:  
o  Plot the random walk. o  
Calculate and plot the rolling mean 
with a window size of 10. o  
crosses above and below zero.  

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 1. Simulate a random walk (cumulative sum of random steps) for 1000 steps, starting at 0
np.random.seed(42)  # Optional: for reproducibility
steps = np.random.choice([-1, 1], size=1000)  # Random steps of -1 or 1
random_walk = np.cumsum(steps)  # Cumulative sum for random walk

# Plot the random walk
plt.figure(figsize=(10, 6))
plt.plot(random_walk, label="Random Walk")
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)  # Add a horizontal line at y=0
plt.title("Random Walk (1000 Steps)")
plt.xlabel("Steps")
plt.ylabel("Position")
plt.legend()
plt.show()

# 2. Calculate and plot the rolling mean with a window size of 10
rolling_mean = pd.Series(random_walk).rolling(window=10).mean()

# Plot the random walk with rolling mean
plt.figure(figsize=(10, 6))
plt.plot(random_walk, label="Random Walk")
plt.plot(rolling_mean, label="Rolling Mean (Window=10)", color='orange', linewidth=2)
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)  # Horizontal line at y=0
plt.title("Random Walk with Rolling Mean (Window Size 10)")
plt.xlabel("Steps")
plt.ylabel("Position")
plt.legend()
plt.show()

# 3. Count the number of times the random walk crosses above and below zero
# Check where the sign of the random walk changes from positive to negative or vice versa
crosses_above_zero = np.sum((random_walk[:-1] <= 0) & (random_walk[1:] > 0))
crosses_below_zero = np.sum((random_walk[:-1] >= 0) & (random_walk[1:] < 0))

print(f"Number of times the random walk crosses above zero: {crosses_above_zero}")
print(f"Number of times the random walk crosses below zero: {crosses_below_zero}")
```

Missing Data Handling  
Find the number of times the walk 
•  Generate a 10x10 matrix of random integers from 1 to 100. Randomly assign NaN values 
to 10% of the matrix:  
o  Count the number of NaN values. o  Replace NaN values with 
the mean of their respective columns. o  Create a mask to filter 
out any rows that contain NaN values. 

```python
import numpy as np
import pandas as pd

# 1. Generate a 10x10 matrix of random integers from 1 to 100
np.random.seed(42)  # For reproducibility
matrix = np.random.randint(1, 101, size=(10, 10)).astype(float)  # Convert to float to allow NaN values

# 2. Randomly assign NaN values to 10% of the matrix
num_nan_values = int(0.1 * matrix.size)  # 10% of the matrix size
nan_indices = (np.random.randint(0, 10, num_nan_values), np.random.randint(0, 10, num_nan_values))
matrix[nan_indices] = np.nan

print("Matrix with NaN values:\n", matrix)

# 3. Count the number of NaN values
nan_count = np.isnan(matrix).sum()
print("\nNumber of NaN values:", nan_count)

# 4. Replace NaN values with the mean of their respective columns
# Use Pandas to handle NaN values more easily
df = pd.DataFrame(matrix)
df.fillna(df.mean(), inplace=True)  # Replace NaN with column mean

print("\nMatrix after replacing NaN values with column mean:\n", df.values)

# 5. Create a mask to filter out any rows that contain NaN values
# (After replacement, there should be no NaN values, but let's create the mask before replacement as an example)
rows_with_nan = df.isna().any(axis=1)  # Mask of rows with NaN values
filtered_matrix = matrix[~rows_with_nan]  # Filter out rows containing NaN

print("\nMatrix with rows containing NaN values removed:\n", filtered_matrix)
```

```python
import numpy as np

# 1. Create a 3D NumPy array of shape (4, 4, 4) filled with random numbers
np.random.seed(42)  # For reproducibility
array_3d = np.random.rand(4, 4, 4)

print("Original 3D array:\n", array_3d)

# 2. Find the sum along each axis
sum_axis0 = np.sum(array_3d, axis=0)  # Sum along axis 0
sum_axis1 = np.sum(array_3d, axis=1)  # Sum along axis 1
sum_axis2 = np.sum(array_3d, axis=2)  # Sum along axis 2

print("\nSum along axis 0:\n", sum_axis0)
print("\nSum along axis 1:\n", sum_axis1)
print("\nSum along axis 2:\n", sum_axis2)

# 3. Swap axes and reshape the array into a 2D matrix
swapped_axes = np.swapaxes(array_3d, 0, 2)  # Swap axis 0 and 2
reshaped_2d = swapped_axes.reshape(16, 4)  # Reshape into a 2D matrix of shape (16, 4)

print("\nArray after swapping axes and reshaping into 2D matrix:\n", reshaped_2d)

# 4. Sort the array along the first axis (axis=0)
sorted_array = np.sort(array_3d, axis=0)

print("\nArray sorted along the first axis:\n", sorted_array)
```

11. Eigenvalues and Eigenvectors  
•  Create a 4x4 symmetric matrix and find its:  
o  Eigenvalues. o  
Eigenvectors. o  
Verify that the matrix 
multiplication of the eigenvectors and their corresponding eigenvalues reconstructs 
the original matrix. 

```python
import numpy as np

# 1. Create a 4x4 symmetric matrix
A = np.array([[4, 1, 2, 3],
              [1, 5, 4, 2],
              [2, 4, 6, 1],
              [3, 2, 1, 7]])

# Ensure the matrix is symmetric
A = (A + A.T) / 2

print("Symmetric matrix A:\n", A)

# 2. Find the eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(A)

print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)

# 3. Verify matrix reconstruction: A ≈ V * D * V.T
# D is the diagonal matrix of eigenvalues
D = np.diag(eigenvalues)

# Matrix multiplication to reconstruct the original matrix
A_reconstructed = eigenvectors @ D @ eigenvectors.T

print("\nReconstructed matrix A from eigenvalues and eigenvectors:\n", A_reconstructed)

# 4. Check if the original matrix is approximately equal to the reconstructed matrix
is_reconstructed = np.allclose(A, A_reconstructed)
print("\nIs the original matrix approximately equal to the reconstructed matrix?:", is_reconstructed)
```
